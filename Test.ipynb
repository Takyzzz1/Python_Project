```python
import requests
import pandas as pd

# Gọi API
url = "https://restcountries.com/v3.1/all?fields=name,latlng,cca3,cca2,region"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

# Parse JSON
data = response.json()

# Chuyển sang DataFrame
records = []
for country in data:
    if "name" in country and "latlng" in country:
        name = country["name"]["common"]
        lat, lon = country["latlng"]
        code= country["cca3"]
        iso_2 = country["cca2"]
        region= country["region"]
        records.append({
            "country": name,
            "latitude": lat,
            "longitude": lon,
            "code": code,
            "iso_2":iso_2,
            "region": region
        })

df_5 = pd.DataFrame(records)
display(HTML(f"""
<div style="overflow-x: auto; overflow-y: auto; height:400px; width:100%;">
{df_5.to_html()}
</div>
"""))


```

```python
GEOJSON_URL = "https://raw.githubusercontent.com/datasets/geo-countries/master/data/countries.geojson"
geojson = requests.get(GEOJSON_URL).json()
iso_list = [feat["properties"]["ISO3166-1-Alpha-3"] for feat in geojson["features"]]
df_4 = pd.DataFrame({"ISO_A3": iso_list, "val": 1})
display(HTML(f"""
<div style="overflow-x: auto; overflow-y: auto; height:400px; width:100%;">
{df_4.to_html()}
</div>
"""))
```

```python
import sqlite3
import pandas as pd

# Kết nối đến database
conn = sqlite3.connect("worldbank.db")

# Lấy danh sách tất cả bảng trong DB
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("📋 Các bảng trong DB:")
print(tables)

# Đọc toàn bộ dữ liệu của từng bảng
for table_name in tables["name"]:
    print(f"\n===== Bảng: {table_name} =====")
    df = pd.read_sql_query(f"SELECT * FROM {table_name};", conn)
    display(df)  # Jupyter sẽ hiển thị đẹp dạng bảng

conn.close()
```

```python
import sqlite3
import pandas as pd
import json

# Kết nối đến database
conn = sqlite3.connect("worldbank.db")
countries = pd.read_sql("SELECT * FROM countries JOIN country_data ON countries.iso2_code = country_data.country_code", conn)

# Giữ dòng có year lớn nhất cho mỗi (iso2_code, indicator_code)
idx = countries.groupby(['iso2_code', 'indicator_code'])['year'].idxmax()
countries = countries.loc[idx][["iso_code", "iso2_code", "name", "indicator_code", "value"]]

# Tạo dict mapping iso_code -> {indicator_code: value}, làm tròn đến 2 chữ số thập phân
indicator_map = countries.groupby('iso_code').apply(
    lambda g: {k: (None if pd.isna(v) else round(float(v), 2)) for k, v in zip(g['indicator_code'], g['value'])}
).to_dict()

# Tạo json_data
json_data = (
    countries[["iso_code", "iso2_code", "name"]]
    .drop_duplicates()
    .set_index("iso_code")
    .assign(indicator=lambda df: df.index.map(lambda iso: indicator_map.get(iso, {})))
    .reset_index()
    .to_dict(orient="records")
)

# In json_data
print(json.dumps(json_data, ensure_ascii=False, indent=2))
```
